```python
from keras.datasets import cifar10
# Завантаження та розділення датасету cifar_10 на тренувальний та тестовий набори
# Нормалізація даних для швидшого та ефективнішого тренування мережі
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
from keras.models import Sequential
from keras.layers import Dense
# Створення нейронної мережі з використанням Keras з чотирма повністю з'єднаними шарами
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(32 * 32 * 3,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
# Запуск тренування моделі на тренувальному наборі
history = model.fit(x_train.reshape(-1, 32 * 32 * 3), y_train, epochs=20, validation_data=(x_test.reshape(-1, 32 * 32 * 3), y_test))
# Виведення графіків функції втрат та точності відносно епох тренування
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.legend()
plt.show()

plt.plot(history.history['accuracy'], label='train_accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.legend()
plt.show()

# Запускаємо тренування моделі
history = model.fit(x_train.reshape(-1, 32 * 32 * 3), y_train, epochs=20, validation_data=(x_test.reshape(-1, 32 * 32 * 3), y_test))

# Виводимо графіки функції втрат та точності
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

plt.plot(history.history['accuracy'], label='train_accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Оцінюємо точність моделі на тестовому наборі
test_loss, test_accuracy = model.evaluate(x_test.reshape(-1, 32 * 32 * 3), y_test)
print(f'Test Loss: {test_loss}')
print(f'Test Accuracy: {test_accuracy}')

import numpy as np

# Вибираємо 10 випадкових зображень з тестового набору
random_indices = np.random.choice(len(x_test), size=10, replace=False)
sample_images = x_test[random_indices]

# Використовуємо модель для класифікації
predictions = model.predict(sample_images.reshape(-1, 32 * 32 * 3))

# Виводимо спрогнозовані класи та порівнюємо їх із справжніми класами
for i in range(10):
    predicted_class = np.argmax(predictions[i])
    true_class = y_test[random_indices[i]]
    print(f"Predicted class: {predicted_class}, True class: {true_class}")


from sklearn.metrics import confusion_matrix, classification_report

# Отримуємо прогнозовані класи для всього тестового набору
all_predictions = model.predict(x_test.reshape(-1, 32 * 32 * 3))
all_predictions_classes = np.argmax(all_predictions, axis=1)

# Виводимо матрицю плутанини та звіт про класифікацію
conf_matrix = confusion_matrix(y_test, all_predictions_classes)
print("Confusion Matrix:")
print(conf_matrix)

class_report = classification_report(y_test, all_predictions_classes)
print("Classification Report:")
print(class_report)

```
